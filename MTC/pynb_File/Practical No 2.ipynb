```python
from mpl_toolkits import mplot3d
import numpy as np
from pylab import*
def f(x,y):
    return np.exp(x**2+y**2)
x=np.linspace(0,2*pi,30)
y=np.linspace(0,2*pi,30)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
ax=axes(projection='3d')
ax.plot_wireframe(X,Y,Z,rstride=2,cstride=2)
xlabel('x')
ylabel('y')
title('exp(x**2+y**2)')
show()
```
